```python
import pandas as pd

# 1. Re-create the sample data (just to make sure we have it ready today)
data = [
    {"Date": "2024-01-01", "Category": "Income", "Description": "Monthly Salary", "Amount": 150000, "Type": "Income"},
    {"Date": "2024-01-02", "Category": "Food", "Description": "Keells Supermarket", "Amount": 12500, "Type": "Expense"},
    {"Date": "2024-01-03", "Category": "Food", "Description": "Lunch Packet", "Amount": 450, "Type": "Expense"},
    {"Date": "2024-01-04", "Category": "Transport", "Description": "PickMe", "Amount": 450, "Type": "Expense"},
    {"Date": "2024-01-25", "Category": "Bills", "Description": "Dialog Bill", "Amount": 2100, "Type": "Expense"},
    {"Date": "2024-01-01", "Category": "Rent", "Description": "Boarding Fees", "Amount": 25000, "Type": "Expense"},
    {"Date": "2024-01-12", "Category": "Shopping", "Description": "Clothes", "Amount": 3500, "Type": "Expense"},
]
df = pd.DataFrame(data)

# 2. READ the data (The Analysis part starts here)
print("--- üìä DATA LOADED ---")
print(df)
```

```python
# 1. Check for missing information
print("--- üßπ CHECKING FOR ERRORS ---")
print(df.isnull().sum())

print("\n--- ‚ÑπÔ∏è DATA TYPES ---")
print(df.dtypes)
```

```python
# 1. Filter only Expenses (Ignore Income for now)
expenses = df[df["Type"] == "Expense"]

# 2. Group by 'Category' and sum the 'Amount'
category_totals = expenses.groupby("Category")["Amount"].sum()

print("--- üí∞ EXPENSES BY CATEGORY ---")
print(category_totals)
```

```python
# 1. Sort the expenses from High to Low
sorted_expenses = expenses.sort_values(by="Amount", ascending=False)

# 2. Get the top item
top_expense = sorted_expenses.iloc[0]

print("--- üö® MOST EXPENSIVE ITEM ---")
print(f"Item: {top_expense['Description']}")
print(f"Cost: Rs. {top_expense['Amount']}")
```
