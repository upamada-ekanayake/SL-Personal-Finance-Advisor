```python
# 1. Defining variables
salary = 150000        # Monthly Income in LKR
rent = 45000           # Boarding / Rent
internet = 2500        # Dialog/SLT Fiber bill
food = 30000           # Supermarket & Eating out

# 2. Doing the math
total_expenses = rent + internet + food
balance = salary - total_expenses

# 3. Showing the result
print("My Total Expenses: Rs.", total_expenses)
print("Money Left (Balance): Rs.", balance)
```

```python
# 1. Ask the user for their input
print("--- SRI LANKA BUDGET CALCULATOR ---")

# We wrap input() inside int() to convert text to a number immediately
salary = int(input("Enter your Monthly Salary (LKR): "))
rent = int(input("Enter your Rent/Boarding fee (LKR): "))
food = int(input("Enter your Food cost (LKR): "))

# 2. Calculate
total_expenses = rent + food
balance = salary - total_expenses

# 3. Show the final report
print("---------------------------------")
print("Total Expenses: Rs.", total_expenses)
print("Remaining Balance: Rs.", balance)
```

```python
# 1. Ask for income and expenses
income = int(input("Monthly Income (LKR): "))
expenses = int(input("Total Monthly Expenses (LKR): "))

# 2. Calculate savings
savings = income - expenses
print("Your Savings: Rs.", savings)

# 3. The Logic (The AI part)
# Calculate 20% of income to compare
target_savings = income * 0.20

if savings < target_savings:
    print("⚠️ ALERT: You are saving less than 20%!")
    print("Try to cut down on non-essential costs.")
else:
    print("✅ GREAT JOB: You are saving more than 20%.")
    print("Consider investing in a Fixed Deposit or Unit Trust.")
```
